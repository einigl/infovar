```python
import numpy as np
import matplotlib.pyplot as plt

from infovar.stats import entropy_estimators as ee
```

```python
N = 10_000

# c1 = np.random.beta(0.5, 0.5, size=N)
# c2 = 0.5 * np.random.normal(0, 1, size=N)

# c1 = np.random.normal(0, 1, size=N)
# c2 = 0.5 * np.random.normal(0, 1, size=N)

c1 = np.random.lognormal(0, 1, size=N)
c2 = 0.5 * np.sign(np.random.normal(0, 1, size=N)) * np.random.lognormal(0, 1, size=N)

# c1 = np.random.triangular(-1, 0, 1, size=N)
# c2 = 0.5 * np.random.triangular(-1, 0, 1, size=N)

# c1 = np.random.uniform(-1, 1, size=N)
# c2 = 0.2 * np.random.uniform(-1, 1, size=N)

x = 0.5 * c1 + 0.5 * c2
y = 0.5 * c1 - 0.5 * c2

rho = np.corrcoef(x, y)[0, 1]
print(rho)
```

```python
plt.hist(x, bins=50)
plt.show()
```

```python
plt.figure(dpi=125)

plt.scatter(x, y)
plt.gca().set_aspect("equal")

plt.show()
```

```python
mi = ee.mi(x.reshape(-1, 1), y.reshape(-1, 1))
lin_mi = - 0.5 * np.log2(1 - rho**2)

print(mi, lin_mi)
```

Il semble qu'une estimation sous hypothèse gaussienne sous-estime systématiquement
